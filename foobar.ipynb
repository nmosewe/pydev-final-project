```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set the plots to display in the Jupyter notebook
%matplotlib inline

# change plotting colors per client request
plt.style.use('ggplot')

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14

df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})

print(df)
```

# Plotting two dataframe columns as a scatter plot

```
# a scatter plot comparing num_children and num_pets
df.plot(kind='scatter',x='num_children',y='num_pets',color='red')
plt.show()
```

## Plot `num_children` vs `age` as scatter plot

```
df.plot(kind='scatter',x='num_children',y='age',color='red')
plt.show()
```

## Plot column values as bar chart

```
df.plot(kind='bar',x='name',y='age')
plt.show()
```

## Plot `gender` vs `age`

```
df.plot(kind='bar',x='gender', y='age')
plt.show()
```

## Plot grouped gender vs age

```
df.groupby('gender').plot(kind='bar',x='gender', y='age')
plt.show()
```

## Plot multilined charts

```
# gca stands for 'get current axis'
ax = plt.gca()

df.plot(kind='line',x='name',y='num_children', color='blue', ax=ax)
df.plot(kind='line',x='name',y='num_pets', color='red', ax=ax)

plt.show()
plt.savefig('output.png')
```

## Bar plot with `group_by`

```
df.groupby('state')['name'].nunique().plot(kind='bar')
plt.show()
```

## Bar plot with `group_by` `gender` and `state`

```
df.groupby(['gender', 'state']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

## Stacked bar plot with two-level group by

```
df
```

```
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

## Changing colors!

```
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True, color=['r', 'g'])
plt.show()
```

## Count the people by gender, spliting by state:

```
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

## Histograms

```
df[['age']].plot(kind='hist',bins=[0,20,40,60,80,100],rwidth=0.8)
plt.show()
```

## Date Histogram

```
df = pd.DataFrame({
    'name':[
        'john','lisa','peter','carl','linda','betty'
    ],
    'date_of_birth':[
        '01/21/1988','03/10/1977','07/25/1999','01/22/1977','09/30/1968','09/15/1970'
    ]
})

df
```

```
def get_month(d):
  return d.month

df['date_of_birth'] = pd.to_datetime(df['date_of_birth'],infer_datetime_format=True)

plt.clf()
df['date_of_birth'].map(get_month).plot(kind='hist')
plt.show()
```
